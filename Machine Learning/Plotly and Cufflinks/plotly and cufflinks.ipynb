```python
df=pd.DataFrame(np.random.randn(100,3),columns=['A','B','C'])
df.head()
```

```python
df.shape
```

```python
df['A']=df['A'].cumsum()+5
df['B']=df['B'].cumsum()+5
df['C']=df['C'].cumsum()+5
df.head()
```

```python
df.iplot()
```

```python
plt.plot(df)
```

```python
df.iplot(x='A',y='B', mode='markers', size=20)
```

```python
cf.datagen.sinwave(10,0.25).iplot(kind='surface')
```

```python
cf.datagen.scatter3d(2,150,mode='stocks').iplot(kind='scatter3d', x='x', y='y', z='z')
```

```python
df[['A', 'B']].iplot(kind='spread')
```

```python
df.iplot(kind='hist', bins = 35, barmode = 'overlay', bargap=0.5)
```

```python
cf.datagen.bubble3d(5,4, mode='stocks').iplot(kind='bubble3d', x='x', y='y', z='z', size='size')
```

```python
ccc=cf.datagen.heatmap(20,20)
ccc
```

```python
ccc.iplot(kind='heatmap', colorscale='spectral', title='Cufflinks - Heatmap')
```
